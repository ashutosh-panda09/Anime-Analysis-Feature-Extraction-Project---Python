```python
import numpy as np 
import pandas as pd 

from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
df = pd.read_csv('anime.csv')
```

```python
df.head()
```

```python
df.loc[0]['Title']
```

```python
def extract_eps(txt):
    ep = ""
    check = False
    for i in txt:
        if i == ")":
            break
        if check == True:
            ep += i
        if i == "(": 
            check = True
     
    return ep
```

##### make a new column for episode count

```python
df['Episodes'] = df['Title'].apply(extract_eps)
```

```python
df['Episodes'] = df['Episodes'].str.replace(" eps", "")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

##### make a new column for time stamp

```python
df.iloc[0]['Title']
```

```python
def extract_time(txt): 
    time_stamp = ''
    for i in range(len(txt)): 
        if txt[i] == ')':
            for j in range(i+1,i+20):
                time_stamp += txt[j]
    return time_stamp
```

```python
df['Total_Time'] = df['Title'].apply(extract_time)
df.head()
```

```python
def extract_months(txt):
    start_txt, end_txt = txt.split(" - ")

    start = datetime.strptime(start_txt, "%b %Y")
    end = datetime.strptime(end_txt, "%b %Y")

    diff = relativedelta(end, start)

    total_months = diff.years * 12 + diff.months

    return total_months


df["Total_Months"] = df["Total_Time"].apply(extract_months)
```
